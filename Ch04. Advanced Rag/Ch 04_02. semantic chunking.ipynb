```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai.embeddings import OpenAIEmbeddings
Text = """
암호화폐 투자는 디지털 자산에 자금을 투자하여 수익을 얻는 방식입니다. 비트코인과 이더리움 같은 주요 암호화폐는 블록체인 기술을 기반으로 하며, 가격 변동성이 크지만 높은 수익을 추구할 수 있습니다. 비트코인, 이더리움과 같은 탈중앙화 금융(DeFi)은 전통 금융 시스템을 대체할 수 있는 기술로, 스마트 계약을 통해 탈중앙화된 금융 서비스를 제공합니다. NFT 기술의 등장으로 이들 암호화폐는 더욱 높은 관심을 받고 있습니다. 지속 가능한 농업은 환경을 보호하면서 안정적인 수확을 목표로 합니다. 스마트 농업은 IoT와 데이터 분석 기술을 활용하여 농작물의 생산성을 극대화합니다. 최근 등장한 스마트팜 기술은 특히 농작물 생산성 향상에 큰 도움을 주고 있습니다.
"""
```

```python
# SemanticChunker 생성
text_splitter = SemanticChunker(OpenAIEmbeddings(model = "text-embedding-3-large"), breakpoint_threshold_type='percentile', breakpoint_threshold_amount=90) # breakpoint_threshold_type='percentile': 의미적 차이의 분포에서 분할 지점 결정 방식 선택
# semantic chunks들로 텍스트 분할
docs = text_splitter.create_documents([Text])
print(docs)
```
